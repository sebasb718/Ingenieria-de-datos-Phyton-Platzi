# Solicitudes a la web 

```python
import requests

response = requests.get('https://www.platzi.com')

print(dir(response))
```

# Extraccin de informacin del HTML 

```python
import bs4

response.encoding = 'utf-8'
soup = bs4.BeautifulSoup(response.text, 'html.parser')

courses_links = soup.select('.RecentCourses-item > a')
courses = [course['href'] for course in courses_links]

for course in courses:
    print(course)
```

```python
import bs4
import requests

response = requests.get('https://www.eltiempo.com')
response.encoding = 'utf-8'

soup = bs4.BeautifulSoup(response.text, 'html.parser')

articles_containers = soup.select('.title-container a')
news = {}
for article in articles_containers:
    news[article.text] = article['href']
    
news
    
```
